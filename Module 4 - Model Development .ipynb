**Linear Regression and Multiple Regression**

```python
from sklearn.linear_model import LinearRegression
import numpy as np


X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.dot(X, np.array([1, 2])) + 3

reg = LinearRegression().fit(X, y)

reg.score(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(np.array([[3, 5]]))
```

From the parameters given by coef_ and intercept_, the linear regression equation is:

$$ y = 3.0000000000000018+ 1.0x_1+ 2.0x_2 $$

**Model Evaluation Using Visualization**


***Regression Plot (regplot)*** shows us a combination of:
    
    * The Scatterplot: where each poit represents a different y;
    * The fitted linear regression line (Å·). 

```python
import seaborn as sns; sns.set(color_codes=True)
import pandas as pd
tips = sns.load_dataset("tips")
ax = sns.regplot(x="total_bill", y="tip", data=tips)

```

***Residual Plot*** represents the error between the actual value:
    
    

```python
df = pd.DataFrame({ 
    'X':np.random.randn(60),
    'Y':np.random.randn(60),
    })

sns.residplot('X','Y',data=df)  
```

As we can see above, the samples are plotted evenly (randomly), so a linear regression is appropriate.
